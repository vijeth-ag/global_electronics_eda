```python
DATASET_PATH = "../dataset"
```

```python
customer_df1 = pd.read_csv(DATASET_PATH+"/Customers.csv", encoding="latin1")
exchange_rates_df1 = pd.read_csv(DATASET_PATH+"/Exchange_Rates.csv", encoding="latin1")
products_df1 = pd.read_csv(DATASET_PATH+"/Products.csv", encoding="latin1")
sales_df1 = pd.read_csv(DATASET_PATH+"/Sales.csv", encoding="latin1")
stores_df1 = pd.read_csv(DATASET_PATH+"/Stores.csv", encoding="latin1")
```

```python
# create new column age using brithday column
customer_df1['Birthday'] = pd.to_datetime(customer_df1['Birthday'], format='%m/%d/%Y')

def calculate_age(born):
    today = datetime.today()
    return today.year - born.year - ((today.month, today.day) < (born.month, born.day))

# Create Age column
customer_df1['age'] = customer_df1['Birthday'].apply(calculate_age)


customer_df1 = customer_df1.drop(['Birthday'], axis=1)

customer_df1.head()
```

```python
from sqlalchemy import create_engine
import pandas as pd

# Replace with your actual MySQL credentials
MYSQL_USER = 'user'
MYSQL_PASSWORD = 'password'
MYSQL_HOST = 'localhost'  # Or your Docker container's IP address
MYSQL_DATABASE = 'my_database'

# Create a connection string
connection_string = f"mysql+pymysql://{MYSQL_USER}:{MYSQL_PASSWORD}@{MYSQL_HOST}/{MYSQL_DATABASE}"

# Create a database engine
engine = create_engine(connection_string)


customer_df1.to_sql('customer', engine, if_exists='replace', index=False)
exchange_rates_df1.to_sql('exchange_rates', engine, if_exists='replace', index=False)
products_df1.to_sql('products', engine, if_exists='replace', index=False)
sales_df1.to_sql('sales', engine, if_exists='replace', index=False)
stores_df1.to_sql('stores', engine, if_exists='replace', index=False)

```
